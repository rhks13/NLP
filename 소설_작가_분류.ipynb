```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words_list = stopwords.words('english')
stop_words_list
```

```python
df.shape[0]
```

```python
def remove_stopwords(text): 
  final_text = []
  for i in text.split():
    if i.strip().lower() not in stop_words_list:
      final_text.append(i.strip())
  return " ".join(final_text)
```

```python
df['text'] = df['text'].str.lower().apply(remove_stopwords)
```

```python
df1['text'].str.lower().apply(alpha_num).apply(remove_stopwords)
```

```python
X_train = np.array([x for x in df['text']])
X_test = np.array([x for x in df1['text']])
y_train = np.array([x for x in df['author']])
```

```python
vocab_size = 20000
embedding_dim = 16
max_length = 500
padding_type='post'
```

```python
import pandas as pd
import warnings 
warnings.filterwarnings(action='ignore')
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
import re
```

```python
tokenizer = Tokenizer(num_words = vocab_size)
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index
```

```python
train_sequences = tokenizer.texts_to_sequences(X_train)
train_padded = pad_sequences(train_sequences, padding=padding_type, maxlen=max_length)

test_sequences = tokenizer.texts_to_sequences(X_test)
test_padded = pad_sequences(test_sequences, padding=padding_type, maxlen=max_length)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])
```

```python
# compile model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# model summary
print(model.summary())
```

```python
num_epochs = 20
history = model.fit(train_padded, y_train, 
                    epochs=num_epochs, verbose=2, 
                    validation_split=0.2)
```
