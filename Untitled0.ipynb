```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
okt = Okt()
okt.morphs('아더빙진짜짜증나네요목소리', stem=True)
```

```python
X_train = []

for sentence in tqdm(train_data['document']):
  tokenized_sentence = okt.morphs(sentence, stem=True)
  stopwords_re = [word for word in tokenized_sentence if not word in stopwords]
  X_train.append(stopwords_re)
```
